```python
from transformers import pipeline
from sklearn.model_selection import train_test_split

```

```python
classifier = pipeline("zero-shot-classification",model = "facebook/bart-large-mnli")
```

```python
sentence_to_classify = "One day I will see the world"

candidate_labels = ["travel", "dancing", "singing"]
```

```python
classifier(sentence_to_classify, candidate_labels)
```

```python
candidate_labels = ["travel", "dancing", "singing", "exploration", "cooking"]
```

```python
classifier(sentence_to_classify, candidate_labels, multi_label=True)
```
